```python
import pandas as pd

earthquakes = pd.read_csv("../../Data/Source/EarthQuakes/earthquakes_2023_global.csv")
```

```python
earthquakes.describe()

```

```python
cord=earthquakes.drop(["dmin","depth"],axis=1)
```

```python
cord
```

```python
cord["nst"]=cord["nst"].fillna(method="bfill")
```

```python
cord["nst"].isnull().sum()
```

```python
cord["gap"]=cord["gap"].fillna(method="ffill")
```

```python
cord["gap"].isnull().sum()
```

```python
cord["gap"]
```

```python
cord["horizontalError"]=cord["horizontalError"].fillna(cord["horizontalError"].mean())
```

```python
cord["horizontalError"].isnull().sum()
```

```python
cord["magError"]=cord["magError"].fillna(cord["magError"].median())
```

```python
cord["magError"].isnull().sum()
```
