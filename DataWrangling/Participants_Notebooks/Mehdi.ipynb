```python
coords = earth_quakes.drop(["latitude","longitude"], axis=1)
coords
```

```python
coords["nst"] = coords["nst"].fillna(method="bfill")
```

```python
coords["nst"] = coords["nst"].fillna(method="ffill")
```

```python
coords["nst"].isnull().sum()
```

```python
coords["magError"]=coords["magError"].fillna(coords["magError"].median())
```
